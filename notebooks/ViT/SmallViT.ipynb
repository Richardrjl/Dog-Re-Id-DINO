```python
import math
import numpy as np
import keras
from keras import ops
from keras import layers
import tensorflow as tf
import matplotlib.pyplot as plt
import os

SEED = 42
keras.utils.set_random_seed(SEED)
```

```python
#Parámetros

BASE_PATH = 'C:/Users/Richard/Desktop/Entorno Tesis/'
PATH = os.path.join(BASE_PATH, 'data/DogFaceNet_224resized/after_4_bis/')       # Ruta a imágenes
PATH_SAVE = os.path.join(BASE_PATH, 'output/history/')

#Configuración de entrenamiento

INPUT_SHAPE = (224, 224, 3)
TEST_SPLIT = 0.1
BATCH_SIZE = 32
```

```python
from tensorflow.preprocessing.image import ImageDataGenerator

# Data generators (con aumento de datos para train)
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    horizontal_flip=True,
    validation_split=TEST_SPLIT  # Divide muestras, no clases
)

test_datagen = ImageDataGenerator(rescale=1./255, validation_split=TEST_SPLIT)

# Flujo de datos desde directorios (estructura: PATH/clase/img.jpg)
train_generator = train_datagen.flow_from_directory(
    PATH,
    target_size=INPUT_SHAPE[:2],
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    subset='training'
)

test_generator = test_datagen.flow_from_directory(
    PATH,
    target_size=INPUT_SHAPE[:2],
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    subset='validation'
)
```
